```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('users2.csv')
```

```python
#df.iloc[:,7:48] = df.iloc[:,7:48].dropna(0)
```

```python
df = df.drop(df[df['Photos12_count'].isnull()].index)
```

```python
df['comments_percent'] = (df['Photos1_count'] + df['Photos2_count'] + df['Photos3_count'] + df['Photos4_count'] + df['Photos5_count'] + df['Photos6_count'] + df['Photos7_count'] + df['Photos8_count'] + df['Photos9_count'] + df['Photos10_count'] + df['Photos11_count'] + df['Photos12_count']) / 12 / df['Followers']
```

```python
df['likes_percent'] = (df['Photos1_likes'] + df['Photos2_likes'] + df['Photos3_likes'] + df['Photos4_likes'] + df['Photos5_likes'] + df['Photos6_likes'] + df['Photos7_likes'] + df['Photos8_likes'] + df['Photos9_likes'] + df['Photos10_likes'] + df['Photos11_likes'] + df['Photos12_likes']) / 12 / df['Followers']
```

```python
df['mean_time'] = 1 / (df['Photos1_time'] - df['Photos2_time'] +
df['Photos2_time'] - df['Photos3_time'] +
df['Photos3_time'] - df['Photos4_time'] +
df['Photos4_time'] - df['Photos5_time'] +
df['Photos5_time'] - df['Photos6_time'] +
df['Photos6_time'] - df['Photos7_time'] +
df['Photos7_time'] - df['Photos8_time'] +
df['Photos8_time'] - df['Photos9_time'] +
df['Photos9_time'] - df['Photos10_time'] +
df['Photos10_time'] - df['Photos11_time'] +
df['Photos11_time'] - df['Photos12_time']) / 11 / 60 / 60 / 24
```

```python
df['rating'] = df['Followers'] * (1 + df['likes_percent']*0.85 + df['comments_percent']*0.15) * (1 + df['mean_time'])
```

```python
df.sort_values(by='rating', ascending=False)
```
